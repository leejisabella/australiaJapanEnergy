# Australia–Japan Energy Investment: Preliminary Analysis

Tests the central hypothesis: Japanese FDI and long-term offtake contracts are positively associated with Australian energy exports to Japan. Two panels are estimated in parallel:

- **Fossil era**: 1962–2024 · dependent variable = coal + LNG export value to Japan (USD)
- **Renewables era**: 2005–2024 · dependent variable = total renewable electricity generation (TWh)

Each regression is run in two forms:
1. **Levels** (matches the model in the project outline) — reported for comparison but biased by non-stationarity.
2. **Log first-differences** — the preferred specification; coefficients read as elasticities of annual growth rates.

Augmented Dickey-Fuller tests are reported up front to motivate the differencing.

## Regression Specification

As outlined in the project proposal, large-scale energy infrastructure such as coal mines and LNG facilities require massive sunk costs that carry severe market risks, including bankruptcy. As a result, this paper hypothesizes that market prices alone are generally insufficient to trigger the levels of capital expenditure required to build these pieces of infrastructure, particularly without guaranteed buyers. Foreign direct investment and long-term offtake agreements are vital governance mechanisms which mitigate the risks of lacking demand, increasing project viability. Thus, a core assumption which underlies this model is that Australian energy output is fundamentally determined by these structural financial commitments, specifically made by Japan, rather than domestic investment alone.

To test the hypothesis, this paper uses a time-series regression model to test Japanese investment and contracts in the fossil fuel era. For the sake of tractability and simplicity, this model assumes a linear relationship between Australian energy export capacity to Japan and several inputs, including foreign direct investment by Japan, the scale of long-term offtake agreements, and Japan's GDP. This model incorporates a time lag to reflect the real-world scenario that these infrastructure projects take several years to build and commission, leading to a multi-year gap before investment is reflected in export capacity. This time lag term uses a variable $k$ to represent the gap in years between investment and energy capacity realization. Based on previous research, this time lag term is set to $k = 3$, reflecting the time required to build significant infrastructure (Edgington 1990). The model of Japanese investment in Australian fossil fuels is expressed as follows:

$$\text{Export Capacity}_t = \beta_0 + \beta_1 (\text{FDI}_{\text{By Japan}})_{t-k} + \beta_2 (\text{Contracts})_t + \beta_3 (\text{GDP}_{\text{Japan}})_t + \varepsilon_t$$

Here, $\text{Export Capacity}_t$ represents Australian energy export capacity to Japan in year $t$, measured as the combined nominal USD value of coal and LNG shipments. The time lag is applied to the Japanese FDI to Australia's energy sector term, to account for construction and scaling after the initial investment has been made; this term is measured in millions of USD. The Contract term is measured as cumulative contracted LNG offtake volume committed to Japanese buyers, in million tonnes per annum (Mtpa). Finally, Japanese GDP serves as a control variable for aggregate baseline demand of energy from Japan, in current USD. If the hypothesis holds, the FDI and Contract coefficients should both be positive and statistically significant. This would be interpreted as meaning that Japanese investment and energy agreements are associated with higher Australian energy production and export volume.

For the renewable energy transition model, this paper uses a parallel time-series regression. Using the same framework allows this paper to directly compare the two investment scenarios, highlighting whether Japanese investment is scaling Australia's clean energy sector at a similar rate to the 1970s level of investment. A key difference from the fossil fuel model is that $\text{Export Capacity}_t$ now measures Australian renewable electricity generation, in terawatt-hours (TWh), as a precursor to future export capacity in hydrogen and ammonia. The FDI term is again Japanese investment into Australia, measured in millions of USD. The Green Contract term is similarly adjusted to capture long-term renewable offtake commitments where available. Japan's GDP is again included as a baseline for macroeconomic demand. The model is as follows:

$$\text{Export Capacity}_t = \alpha_0 + \alpha_1 (\text{Green FDI}_{\text{By Japan}})_{t-k} + \alpha_2 (\text{Green Contracts})_t + \alpha_3 (\text{GDP}_{\text{Japan}})_t + \varepsilon_t$$

To interpret this model, this paper examines the coefficients of Green FDI and Green Contracts. If both are positive and statistically significant, they will imply that historical financial mechanisms are successfully adapting to the clean energy transition and significantly increasing Australian renewables capacity oriented toward Japanese demand.

Both regressions are estimated in two forms. The first is a **levels** specification, which matches the model as written in the project proposal; this is reported for direct comparison with the proposal. The second is a **log first-difference** specification, in which every continuous variable enters as $\Delta \ln(\cdot)$. This second form is the preferred specification because the underlying macro and trade series are non-stationary in levels (see the Augmented Dickey-Fuller tests below), and running OLS on non-stationary series produces spurious results. In the log-difference form, coefficients are interpreted as elasticities of annual growth rates rather than levels. Heteroscedasticity- and autocorrelation-consistent (HAC) standard errors are reported in both cases to account for serial correlation in the residuals.

## Data Variable Descriptions

As noted in the project proposal, there is no single dataset which can be used to test all parts of this paper's research question. This amalgamation of data into a new dataset demonstrates the novelty of this question and how the subject area is understudied. As such, the panels used below are compiled across multiple authoritative sources. Two panels are constructed in parallel: a **fossil era** panel covering 1962–2024 and a **renewables era** panel covering 2005–2024.

### Dependent variables

- **`export_energy_usd_to_jpn`** — Australian energy export value to Japan in year $t$, measured in current USD and computed as the sum of `export_coal_usd_to_jpn` and `export_lng_usd_to_jpn`. This serves as the primary measure of Australian export capacity to Japan in the fossil fuel model. Source: UN Comtrade and the Australian Bureau of Statistics. The disaggregated coal and LNG series are also retained for descriptive analysis.
- **`renewables_electricity`** — Australian renewable electricity generation in terawatt-hours (TWh), capturing the combined output of hydro, wind, and solar generation. This serves as the dependent variable in the renewables model. Because Australia does not yet export renewable electricity to Japan at scale, this paper uses domestic generation as a proxy for the capacity that will eventually underpin hydrogen and ammonia exports. Source: Our World in Data, drawing on Ember and the Energy Institute. The series is decomposed into `solar_electricity`, `wind_electricity`, and `hydro_electricity` for descriptive analysis, and `renewables_share_elec` reports renewables as a share of total electricity generation.

### Explanatory variables

- **`fdi_flow_usd_million`** — Annual Japanese foreign direct investment flow into Australia, in millions of current USD. This term enters the regression with a $k = 3$ year lag to reflect that infrastructure projects require multiple years between initial investment and the realization of export capacity. Source: DFAT International Investment Australia tables and OECD FDI statistics.
- **`fdi_stock_usd_million`** — Cumulative Japanese FDI stock held in Australia, in millions of current USD. This variable is preferred over annual flow in the renewables-era panel because investment in renewable projects is recent and lumpy, making stock a more stable indicator of accumulated capital commitment. Source: DFAT International Investment Australia tables.
- **`japan_lng_contracted_mtpa`** — Cumulative volume of Australian LNG committed to Japanese buyers under long-term offtake agreements, in million tonnes per annum (Mtpa). This proxies the Contracts term in the fossil fuel regression and captures the institutional mechanism through which Japanese demand "secured" Australian energy resources beyond pure spot-market exchange. Source: hand-compiled from project disclosures including the North West Shelf Venture, Darwin LNG, Pluto LNG, Gorgon, Wheatstone, and Ichthys LNG, supplemented by the Institute for Energy Economics and Financial Analysis's Australian Gas and LNG Tracker.
- **`jpn_gdp_current_usd`** — Japanese GDP in current USD, included as a control for aggregate baseline Japanese demand for imported energy. This isolates the effect of FDI and contracts from broader macroeconomic expansion in Japan. Source: World Bank World Development Indicators.

### Control variables

- **`jpy_per_aud`** — Annual average exchange rate of Japanese yen per Australian dollar, included to control for relative price movements that affect the USD-denominated value of bilateral trade independent of physical volume. Source: Reserve Bank of Australia.
- **`coal_australian_usd_per_mt`** — World Bank Pink Sheet benchmark price for Australian thermal coal, in current USD per metric tonne. This controls for the commodity price channel through which export value can rise without any corresponding change in investment or contracted volume. Source: World Bank Commodity Markets ("Pink Sheet").
- **`lng_japan_usd_per_mmbtu`** — World Bank Pink Sheet benchmark Japan LNG import price, in current USD per million British thermal units. As with the coal price, this controls for the price channel in the fossil fuel model and additionally serves as a proxy for the relative cost of dirty energy in the renewables specification.

### Transformations

Because most macroeconomic and trade series are integrated of order one (see the ADF table in Section 2), each continuous variable is also constructed in log first-difference form (denoted `dl_<var>`). The Contract variable enters as a simple first difference rather than a log-difference because new contract volumes are often zero in a given year, which is undefined under the log transformation. The lagged FDI term `dl_fdi_lag` applies the three-year lag $k$ after taking the log-difference, preserving the construction-period interpretation in the differenced specification.

```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller

REPO = Path.cwd().parent
PROC = REPO / "data" / "processed"

fossil = pd.read_csv(PROC / "fossil_era_panel.csv")
renew = pd.read_csv(PROC / "renewables_era_panel.csv")
fossil.shape, renew.shape
```

## 2. Stationarity checks (ADF)
Most macro/trade series are I(1). If we cannot reject a unit root, OLS on levels yields spurious results.

```python
def adf_table(df, cols):
    rows = []
    for c in cols:
        x = df[c].dropna()
        if len(x) < 10:
            rows.append({"series": c, "n": len(x), "adf_stat": np.nan, "p_value": np.nan})
            continue
        stat, p, *_ = adfuller(x, autolag="AIC")
        rows.append({"series": c, "n": len(x), "adf_stat": stat, "p_value": p})
    return pd.DataFrame(rows)

adf_table(fossil, core).round(3)
```

Series with p > 0.10 are non-stationary in levels — take first differences.

## 3. Fossil-era regression
Two specifications, both with **k = 3** year lag on FDI:

**Levels (outline model):**
$$\text{Exports}_t = \beta_0 + \beta_1 \text{FDI}_{t-3} + \beta_2 \text{LNG\_Contracts}_t + \beta_3 \text{Japan\_GDP}_t + \beta_4 \text{LNG\_price}_t + \beta_5 \text{Coal\_price}_t + \varepsilon_t$$

**Log first-differences (preferred):** every variable enters as $\Delta \ln(\cdot)$ so coefficients are elasticities.

```python
K = 3  # lag on FDI

f = fossil.copy()
f["fdi_lag"] = f["fdi_flow_usd_million"].shift(K)

# Levels regression (1990 onwards so all variables are present)
lev = f[(f.year >= 1990) & (f.year <= 2024)].dropna(
    subset=["export_energy_usd_to_jpn", "fdi_lag", "japan_lng_contracted_mtpa",
            "jpn_gdp_current_usd", "lng_japan_usd_per_mmbtu", "coal_australian_usd_per_mt"]
)
X = lev[["fdi_lag", "japan_lng_contracted_mtpa", "jpn_gdp_current_usd",
         "lng_japan_usd_per_mmbtu", "coal_australian_usd_per_mt"]]
y = lev["export_energy_usd_to_jpn"]
X = sm.add_constant(X)
lev_model = sm.OLS(y, X).fit(cov_type="HAC", cov_kwds={"maxlags": 2})
print(lev_model.summary())
```

```python
# Log first-differences
def dlog(s):
    return np.log(s.replace({0: np.nan})).diff()

g = f.copy()
g["dl_export"]   = dlog(g["export_energy_usd_to_jpn"])
g["dl_fdi_lag"]  = dlog(g["fdi_flow_usd_million"]).shift(K)
g["dl_contract"] = g["japan_lng_contracted_mtpa"].diff()  # in level Mtpa (often 0)
g["dl_gdp"]      = dlog(g["jpn_gdp_current_usd"])
g["dl_lng_p"]    = dlog(g["lng_japan_usd_per_mmbtu"])
g["dl_coal_p"]   = dlog(g["coal_australian_usd_per_mt"])
g["dl_fx"]       = dlog(g["jpy_per_aud"])

reg = g[(g.year >= 1992) & (g.year <= 2024)].dropna(
    subset=["dl_export", "dl_fdi_lag", "dl_contract", "dl_gdp",
            "dl_lng_p", "dl_coal_p", "dl_fx"]
)
X2 = reg[["dl_fdi_lag", "dl_contract", "dl_gdp", "dl_lng_p", "dl_coal_p", "dl_fx"]]
y2 = reg["dl_export"]
X2 = sm.add_constant(X2)
diff_model = sm.OLS(y2, X2).fit(cov_type="HAC", cov_kwds={"maxlags": 2})
print(diff_model.summary())
```

## 5. Renewables-era regression
Because the panel only has 20 observations and the Contracts variable does not yet exist for renewables (hydrogen/ammonia offtake agreements are still nascent), we run a simple specification:
$$\Delta \ln (\text{Renewables\_TWh})_t = \alpha_0 + \alpha_1 \Delta \ln(\text{FDI\_stock})_{t-3} + \alpha_2 \Delta \ln (\text{Japan\_GDP})_t + \alpha_3 \Delta \ln (\text{LNG\_price})_t + \varepsilon_t$$

```python
r = renew.copy()
# Join in LNG price as a proxy for the relative price of dirty energy.
prices = pd.read_csv(REPO / "data" / "raw" / "pink_sheet_prices.csv")
lng = prices[prices.commodity == "lng_japan_usd_per_mmbtu"][["year", "price_nominal_usd"]]
lng = lng.rename(columns={"price_nominal_usd": "lng_price"})
r = r.merge(lng, on="year", how="left")

K = 3
r["dl_ren"]      = dlog(r["renewables_electricity"])
r["dl_fdi_lag"]  = dlog(r["fdi_stock_usd_million"]).shift(K)
r["dl_gdp"]      = dlog(r["jpn_gdp_current_usd"])
r["dl_lng_p"]    = dlog(r["lng_price"])
reg = r.dropna(subset=["dl_ren", "dl_fdi_lag", "dl_gdp", "dl_lng_p"])
print(f"n = {len(reg)}")
X3 = sm.add_constant(reg[["dl_fdi_lag", "dl_gdp", "dl_lng_p"]])
y3 = reg["dl_ren"]
ren_model = sm.OLS(y3, X3).fit(cov_type="HAC", cov_kwds={"maxlags": 1})
print(ren_model.summary())
```

## 6. Comparison
How does the FDI elasticity compare across the two eras?

```python
def coef_row(name, model, key):
    return {
        "era": name,
        "variable": key,
        "coef": model.params.get(key),
        "se":   model.bse.get(key),
        "p":    model.pvalues.get(key),
        "n":    int(model.nobs),
    }

summary = pd.DataFrame([
    coef_row("fossil (Δlog)", diff_model, "dl_fdi_lag"),
    coef_row("renewables (Δlog)", ren_model, "dl_fdi_lag"),
])
summary.round(3)
```

## 7. Caveats & next steps

- The LNG contracts variable proxies long-term commitment volume; coking-coal contracts are not yet included.
- 20 observations in the renewables panel make standard errors loose; add hydrogen project announcements (HESC, Yuri, CQ-H2) when those data are available.
- Consider a structural break test around 1989 (NWS first cargo) and 2011 (Fukushima).
